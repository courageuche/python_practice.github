```python
import numpy as np
import pandas as pd
from numpy import random
#Gaussian / Normal distribution
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.plot(np.linspace(-1, 1, 500),norm.pdf(np.linspace(-1, 1, 500)))
plt.show()
```

```python
part_ = np.random.normal(loc=1, scale = 1, size = 500)
plt.hist(part_)
```

```python
#Bernoullis
from scipy.stats import bernoulli
prob = 0.5
angle = bernoulli(prob)
fig, _axis = plt.subplots(1,1)
x = [0,1]
_axis.vlines(x, 0, angle.pmf(x), label = 'dat')
plt.show()
```

```python
#Binomial
from scipy.stats import binom
prob = 0.3
a = 7
x = np.linspace(0, a, a+1)
angle = binom(a, prob)
fig, _axis = plt.subplots(1,1)
_axis.vlines(x, 0, angle.pmf(x), label = 'dat')
plt.show()
```

```python
#Poisson
from scipy.stats import poisson
y = 3 #lambda
x = np.linspace(0, 12, num=14)
angle = poisson(y)
fig, _axis = plt.subplots(1,1)
_axis.vlines(x, 0, angle.pmf(x), label = 'dat')
plt.show()
```
